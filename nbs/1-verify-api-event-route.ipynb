```python
import requests
```

```python
path= "/api/events"
base_url= "http://localhost:8000"
endpoint = f'{base_url}{path}'
endpoint 
```

```python
response = requests.get(endpoint)
response.status_code
```

```python
if response.ok:
    data = response.json()
    print(f"API is reachable and returned a successful response. Data: {data}")
else:
    print("Failed to reach the API or received an error response.")
```
