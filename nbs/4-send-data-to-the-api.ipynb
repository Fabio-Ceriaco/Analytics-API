```python
import requests
import json

path= "/api/events"
base_url= "http://localhost:8000"
endpoint = f'{base_url}{path}'
endpoint 


headers ={
    'Content-Type': 'application/json'
}
data = {'page': '/test+'}
response = requests.post(endpoint, data=json.dumps(data), headers=headers)
print(response.headers)

if response.ok:
    data = response.json()
    print(f"API is reachable and post was successful. Data: {data}")
else:
    print(response.text)
```
