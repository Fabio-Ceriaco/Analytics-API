```python
import requests
import json
```

```python
create_path= "/api/events"
base_url= "http://localhost:8000"
create_endpoint = f'{base_url}{create_path}'
```

```python
headers ={
    'Content-Type': 'application/json'
}

response = requests.post(create_endpoint, json={'page': 'test+'}) # HTTP Post
print(response.headers)

if response.ok:
    data = response.json()
    print(f"API is reachable and post was successful. Data: {data}")
else:
    print(response.text)
```
