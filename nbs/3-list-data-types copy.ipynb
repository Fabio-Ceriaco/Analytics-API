```python
import requests
from pprint import pprint
```

```python
create_path= "/api/events"
base_url= "http://localhost:8000"
create_endpoint = f'{base_url}{create_path}'
```

```python
headers ={
    'Content-Type': 'application/json'
}

response = requests.post(create_endpoint, json={'page': 'test+'}) # HTTP Post
print(response.headers)

if response.ok:
    data = response.json()
    pprint(f"API is reachable and post was successful. Data: {data}")
else:
    print(response.text)
```

```python
import random

event_count = 10000
pages = ["/home", "/about", "/contact", "/products", "/services", "/blog", "/careers"]

for i in range(event_count):
    page = random.choice(pages)
    response = requests.post(create_endpoint, json={'page': page})
    if response.ok:
        data = response.json()
        print(data, type(data), data.get('items'))
    else:
        print(response.text)
```

```python
update_path= "/api/events/4"
base_url= "http://0.0.0.0:8000"
update_endpoint = f'{base_url}{update_path}'
```

```python
payload = {'description': 'Updated description'}

response = requests.put(update_endpoint, json= payload)

if response.ok:
    data = response.json()
    print(f'update response {data}')
else:
    print('update failed', response.status_code)
```

```python
delete_path = "/api/events/4"
base_url= "http://0.0.0.0:8000"
delete_endpoint = f'{base_url}{delete_path}'
```

```python
response = requests.delete(delete_endpoint)
if response.ok:
    data = response.json()
    print('delete response', data['detail'])
print('delete status', response.status_code)
```
